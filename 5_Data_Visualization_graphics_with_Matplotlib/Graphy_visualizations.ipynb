#Fazendo as primeiras visualizações

```python
import pandas as pd
```

#Espessura, estilo, foco, marcadores e grades

```python
fig = plt.figure(figsize=(15,8))

eixo_maio = fig.add_axes([0,0,1,1])
eixo2014 = fig.add_axes([0.7,0.65,0.3,0.3])

eixo_maio.grid(True)
eixo_maio.plot(df['data'], df['temperatura'], color= 'teal')
eixo_maio.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo_maio.set_title('Temperatura em Maio de 2014', fontsize= 25, pad = 20)
eixo_maio.set_ylabel('Tempetarura em F', fontsize= 20)
eixo_maio.set_xlabel('Data', fontsize= 20)
eixo_maio.legend(['temperatura'], loc= 'lower right', fontsize=15)

eixo2014.plot(df['data'], df['temperatura'], color='b') #b de blue, 
#as principais cores podem ser usadas apenas a primeira letra que funciona ex"r"de red
eixo2014.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2014.set_title('Temperatura em 2014', fontsize= 12)
eixo2014.set_ylabel('Tempetarura em F', fontsize= 10)
eixo2014.set_xlabel('Data', fontsize= 10)
eixo2014.legend(['temperatura'], loc= 'best', fontsize=8)
eixo2014.grid(True)
```

```python
#Alterando os limites

fig = plt.figure(figsize=(15,8))

eixo_maio = fig.add_axes([0,0,1,1])
eixo2014 = fig.add_axes([0.7,0.65,0.3,0.3])

eixo_maio.grid(True)
eixo_maio.plot(df['data'], df['temperatura'], color= 'teal')
eixo_maio.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo_maio.set_ylim(270,320)
eixo_maio.set_title('Temperatura em Maio de 2014', fontsize= 25, pad = 20)
eixo_maio.set_ylabel('Tempetarura em F', fontsize= 20)
eixo_maio.set_xlabel('Data', fontsize= 20)
eixo_maio.legend(['temperatura'], loc= 'lower right', fontsize=15)

eixo2014.plot(df['data'], df['temperatura'], color='b') 
eixo2014.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2014.set_title('Temperatura em 2014', fontsize= 12)
eixo2014.set_ylabel('Tempetarura em F', fontsize= 10)
eixo2014.set_xlabel('Data', fontsize= 10)
eixo2014.legend(['temperatura'], loc= 'best', fontsize=8)
eixo2014.grid(True)
```

```python
#Criando filtro que corresponde a uma área específica do gráfico
#No caso seria uma marcação no mês de maio do graf. 2014

fig = plt.figure(figsize=(15,8))

eixo_maio = fig.add_axes([0,0,1,1])
eixo2014 = fig.add_axes([0.7,0.65,0.3,0.3])

eixo_maio.grid(True)
eixo_maio.plot(df['data'], df['temperatura'], color= 'teal')
eixo_maio.set_xlim(datetime.datetime(2014,5,1),datetime.datetime(2014,6,1))
eixo_maio.set_ylim(270,320) #escala, limites do eixo y
eixo_maio.set_title('Temperatura em Maio de 2014', fontsize= 25, pad = 20)
eixo_maio.set_ylabel('Tempetarura em F', fontsize= 20)
eixo_maio.set_xlabel('Data', fontsize= 20)
eixo_maio.legend(['temperatura'], loc= 'lower right', fontsize=15)

#Criando filtro que corresponde a uma área específica do gráfico
#Teal é a cor do gráfico
blue_esquerda = df['data'] < datetime.datetime(2014,5,1)
blue_direita = df['data'] > datetime.datetime(2014,6,1)

eixo2014.plot(df['data'], df['temperatura'], color='teal') 
eixo2014.plot(df[blue_esquerda]['data'], df[blue_esquerda]['temperatura'], color = 'b')
eixo2014.plot(df[blue_direita]['data'], df[ blue_direita]['temperatura'], color = 'b')
eixo2014.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1))
eixo2014.set_title('Temperatura em 2014', fontsize= 12)
eixo2014.set_ylabel('Tempetarura em F', fontsize= 10)
eixo2014.set_xlabel('Data', fontsize= 10)
eixo2014.legend(['temperatura'], loc= 'best', fontsize=8)
eixo2014.grid(True)
```
