#Fazendo as primeiras visualizações

```python
import pandas as pd
```

```python
df = pd.read_csv('monitoramento_tempo.csv')
```

```python
df.head()
```

```python
df.info() #verifica dados e formatos das classes 
```

```python
df['data'] = pd.to_datetime(df['data']) #convertendo os dados de data que estavam como objeto para dados de data realmente
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
# faz com que os gráficos no jupiter apareçam de forma automática
```

```python
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
```

```python
plt.figure(figsize=(15,8))
plt.plot(df['data'], df['temperatura'])
plt.title('Temperatura no tempo') #inserindo o título
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento')
eixo.set_ylabel('Tempetarura em F')
eixo.set_xlabel('Data')
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'])

eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal') # uma das cores disponíveis
#ver documentação da lista de cores: https://matplotlib.org/stable/gallery/color/named_colors.html

eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

#Espessura, estilo, foco, marcadores e grades

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal', lw = 0.5) #lw altera a espessura da linha

eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal', ls= 'dotted') # ls altera o tipo da linha para pontilhado

eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

```python
import datetime
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal')

# foca em um período específico do gráfico
# para isso é necessário também importar o datetime
eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1)) 
eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal', marker = 'o') 
#inserindo marcador para cada dado de entrada

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1)) 
eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color= 'teal') 

eixo.set_xlim(datetime.datetime(2014,1,1),datetime.datetime(2015,1,1)) 
eixo.set_title('Temperatura no momento', fontsize= 25)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'lower right', fontsize=15)
eixo.grid(True) #inserindo linha de grade
```

```python
fig = plt.figure(figsize=(15,8))
#Inserindo um gráfico dentro de outro gráfico
eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.1,0.5,0.4,0.4])
```

```python
fig = plt.figure(figsize=(15,8))

eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.1,0.5,0.4,0.4])

eixo.grid(True)
eixo.plot(df['data'], df['temperatura'], color= 'teal')
eixo.set_title('Temperatura no momento', fontsize= 25, pad = 20)
eixo.set_ylabel('Tempetarura em F', fontsize= 20)
eixo.set_xlabel('Data', fontsize= 20)
eixo.legend(['temperatura'], loc= 'best', fontsize=15)
```
