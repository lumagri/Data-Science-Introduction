#Dados sobre n√∫mero de pessoas que se recuperaram e pessoas que morreram. 

```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")

display(deaths.head())
display(recovered.head())
#o display √© umafun√ß√£o do colab que funciona igual ao print. 
#Para este tipo de visualiza√ß√£o a seguir ele funciona melhor.
```

#Criando uma fun√ß√£o com as 3 base de dados por pa√≠ses. 

```python
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]

latest_by_country(confirmed)
```

```python
informations = [latest_by_country(confirmed),latest_by_country(deaths),latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
#Precisa arrumar o nome das colunas. 
```

```python
informations = [latest_by_country(confirmed),latest_by_country(deaths),latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

```python
informations = [latest_by_country(confirmed),latest_by_country(deaths),latest_by_country(recovered)]
combined = pd.concat(informations, axis=1)
combined.columns = ['confirmed', 'deaths', 'recovered']
combined.head()
```

```python
#Taxa de letalidade √© expressa em %
# o n√∫mero de casos letais dentro dos casos foram letais e n√£o letais

#imagina que existe uma doen√ßa que em janeiro de 2020 afetou 100 pessoas
# 5 pessoas morreram (por exemplo). letalidade foi de 5%. 
# n√£o pode dizer que ela √© de 5% para sempre pois a doen√ßa continua existindo. 
#Precisa de um padr√£o bem definido para fazer a compara√ß√£o em tempos diferentes. 

#√â dif√≠cil falar de letalidade tentando ser preciso durante a pandemia pois no come√ßo
#√© comum que somente casos graves sejam reportados. 
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up['deaths']/ sum_up['confirmed']
print(round(letality_rate_1 * 100, 2))
```

```python
letality = combined['deaths']/ combined['confirmed'] * 100
combined['letality_rate_1'] = letality_rate_1
combined.head()
```

```python
letality_rate_2 = combined["deaths"] / (combined["deaths"] + combined["recovered"])
```

```python
sum_up = combined.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
letality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"])
print(letality_rate_1 * 100) #Mortes
print(letality_rate_2 * 100) #recuperados
```

```python
def latest_by_country_at(data, date):
  return data.groupby("Country/Region").sum()[date]


informations = [latest_by_country_at(confirmed, '2/20/20'), latest_by_country_at(deaths, '2/20/20'), latest_by_country_at(recovered, '2/20/20')]
combined_2_20_20 = pd.concat(informations, axis=1)
combined_2_20_20.columns = ["confirmed", "deaths", "recovered"]

sum_up = combined_2_20_20.sum()
letality_rate_1 = sum_up["deaths"] / sum_up["confirmed"]
letality_rate_2 = sum_up["deaths"] / (sum_up["deaths"] + sum_up["recovered"])
print(letality_rate_1 * 100)
print(letality_rate_2 * 100)
```
